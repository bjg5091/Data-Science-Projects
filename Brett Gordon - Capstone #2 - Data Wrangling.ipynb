```python
#Game Cell #2

df_game = pd.read_sql_query("SELECT * FROM Game", df2)

df_game = df_game[['GAME_ID', 'SEASON_ID', 'TEAM_ID_HOME', 'TEAM_ABBREVIATION_HOME', 'TEAM_NAME_HOME', 'GAME_DATE', 'MATCHUP_HOME',
                  'WL_HOME', 'PTS_HOME', 'PLUS_MINUS_HOME', 'TEAM_ID_AWAY', 'TEAM_ABBREVIATION_AWAY', 'TEAM_NAME_AWAY',
                  'WL_AWAY', 'PTS_AWAY', 'GAMECODE', 'TEAM_WINS_LOSSES_HOME', 'TEAM_WINS_LOSSES_AWAY']].copy()
print(df_game.head())
```

```python
#Game Cell #3

df_game['year'] = pd.DatetimeIndex(df_game['GAME_DATE']).year
#df_game['GAME_DATE'] = pd.to_datetime(df_game['GAME_DATE'], format='%Y-%m-%d')
df_game = df_game[df_game['year'] >=2014]
#print(df_game)

df_game['PTS_HOME'] = df_game['PTS_HOME'].astype(int)
df_game['PTS_AWAY'] = df_game['PTS_AWAY'].astype(int)
df_game = df_game.assign(Total =lambda x: x.PTS_HOME + x.PTS_AWAY)
df_total = df_game.sort_values(by=['Total','PTS_HOME','PTS_AWAY'])
%store df_total
df_total
```

```python
#Game Cell #4

missing = pd.concat([df_game.isnull().sum(), 100 * df_game.isnull().mean()], axis=1)
missing.columns=['count','%']
missing.sort_values(by=['count','%'])
```

```python
missing_wl = df_game[['WL_HOME', 'WL_AWAY', 'TEAM_WINS_LOSSES_HOME', 'TEAM_WINS_LOSSES_AWAY']].isnull().sum(axis=1)
df_game = df_game[missing_wl == 0]
missing = pd.concat([df_game.isnull().sum(), 100 * df_game.isnull().mean()], axis=1)
missing.columns=['count','%']
missing.sort_values(by=['count','%'])
```

```python
df_game[['home_wins', 'home_losses']] = df_game['TEAM_WINS_LOSSES_HOME'].str.split('-', expand=True).astype(int)
df_game[['away_wins', 'away_losses']] = df_game['TEAM_WINS_LOSSES_AWAY'].str.split('-', expand=True).astype(int)
df_game = df_game.assign(home_pct=lambda x: x.home_wins/(x.home_wins + x.home_losses))
df_game = df_game.assign(away_pct=lambda x: x.away_wins/(x.away_wins + x.away_losses))
df_game = df_game.assign(total_games_played=lambda x: x.home_wins + x.away_wins + x.home_losses + x.away_losses)

%store df_game
df_home_wins = df_game.sort_values(by=['home_pct', 'away_pct', 'home_wins', 'away_wins'])
%store df_home_wins
df_home_wins.tail(100)
```

```python
df_game_40_plus = df_game[df_game['total_games_played'] >= 40]
df_game_40_plus = df_game_40_plus.sort_values(by=['home_pct', 'away_pct', 'home_wins', 'away_wins'])
%store df_game_40_plus
df_game_40_plus.tail(100)
```
